```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression
```

```python
X,Y = make_regression(n_samples = 1000,n_features = 2,n_informative=1,noise = 10,random_state=1)

print(X.shape,Y.shape)
```

```python
plt.subplot(1,2,1)
plt.scatter(X[:,0],Y)

plt.subplot(1,2,2)
plt.scatter(X[:,1],Y)
```

```python
from mpl_toolkits import mplot3d
%matplotlib notebook 
# to make the plot interactive
fig = plt.figure(figsize = (10,7))
ax = plt.axes(projection = '3d')

# Creating plot
ax.scatter3D(X[:,0],X[:,1],Y,color = 'green')
plt.title("3D scatter plot")
plt.show()
```

### Implementing Linear Regression using sklearn

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,Y)
```

```python
model.coef_ # theta1 - thetan
```

```python
model.intercept_ # theta0
```

```python
model.predict([X[0],X[1]])
```

```python
Y[0],Y[1]
```

```python
model.score(X,Y)
```
