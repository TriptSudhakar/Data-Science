```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_regression
```

```python
# Generate dataset
X,Y = make_regression(n_samples = 400,n_features = 1,n_informative = 1,noise = 1.8,random_state = 11)
Y = Y.reshape((-1,1))
print(X.shape)
print(Y.shape)
```

```python
# Normalize
X = (X-X.mean())/X.std()

# Visualise
plt.figure()
plt.scatter(X,Y)
plt.title("Normalized Data")
plt.show()
```

```python
ones = np.ones((X.shape[0],1))
X_ = np.hstack((ones,X))
print(X_.shape)
```

```python
def predict(X,theta):
    return np.dot(X,theta)

def getThetaclosedform(X,Y):
    firstpart = np.dot(X.T,X)
    secondpart = np.dot(X.T,Y)
    
    theta = np.dot(np.linalg.pinv(firstpart),secondpart)
    return theta
```

```python
theta = getThetaclosedform(X_,Y)
print(theta)
```

```python
plt.figure()
plt.scatter(X,Y)
plt.plot(X,predict(X_,theta),color = 'red')
plt.title("Normalized Data")
plt.show()
```
