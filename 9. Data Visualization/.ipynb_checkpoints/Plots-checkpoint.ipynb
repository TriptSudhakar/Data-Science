```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
themes = plt.style.available
print(themes)
```

```python
plt.style.use("seaborn")
```

### Line Plots

```python
x = np.arange(10)
print(x)
```

```python
y1 = x**2
y2 = 2*x+3
print(y1)
print(y2)
```

```python
plt.plot(x,y1,color = 'red',label = "Apple",marker = 'o')
plt.plot(x,y2,color = 'green',label = "Kiwi",linestyle = "dashed",marker='*')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of fruits over time")
plt.legend()
plt.show() # this is the method used to show the output, has to be used in a .py file
# to draw on two different graphs , use plt.show() twice
```

```python
prices = np.array([1,2,3,4])**3
print(prices)
```

```python
plt.plot(prices) # takes the indices as the default arguement over x axis
plt.show() 
```

### Scatter Plots

```python
plt.figure(figsize = (5,5)) # Adjust the size of the plot
plt.scatter(x,y1,color = 'red',label = "Apple",marker = 'o')
plt.scatter(x,y2,color = 'green',label = "Kiwi",linestyle = "dashed",marker='*')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter plot - Prices over time")
plt.legend()
plt.show()
```

### Bar Graphs

```python
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,15],width = 0.5,label = "Current Year",tick_label = ["Gold","Platinum","Silver"])
plt.bar(x_coordinates+0.25,[20,10,12],width = 0.5,label = "Next Year")
plt.title("Metal Price Comparision")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
plt.style.use("dark_background")
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,15],width = 0.5,label = "Current Year",tick_label = ["Gold","Platinum","Silver"],color = "green") 
plt.bar(x_coordinates+0.25,[20,10,12],width = 0.5,label = "Next Year",color = "grey")
plt.title("Metal Price Comparision")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,40)
plt.xlim(-1,5)
plt.legend()
plt.show()
```

### Pie Charts

```python
plt.style.use("seaborn")
```

```python
subjects = "Maths","Physics","Chem","English"
weightages = [20,10,15,5]
plt.pie(weightages,labels = subjects,explode = (0,0,0.1,0),autopct = "%1.1f%%",shadow = True) 
plt.show()
```

### Histogram

```python
Xsn = np.random.randn(100)
```

```python
u = 70
sigma = 8
X = Xsn*sigma + u
X2 = np.round(Xsn*15+40)
```

```python
print(X,X2)
```

```python
X = np.round(X)
print(X)
```

```python
plt.style.use("seaborn")
plt.title("Histogram")
plt.hist(X,alpha = 0.7,label = "Maths")
plt.hist(X2,alpha = 0.8,label = "Physics")
plt.xlabel("Marks range")
plt.ylabel("Prob/Freq Count of students")
plt.show()
```
